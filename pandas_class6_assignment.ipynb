# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
import pandas as pd
info = pd.DataFrame({
    'City':["Bhopal","Indore","Bhopal","Delhi"]
})
print(info['City'].drop_duplicates())
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
info = pd.DataFrame({
    'City':['Bhopal','Bhopal','Delhi','Bhopal','Delhi'],
    'Marks':[80,90,70,85,70]
})
avg_marks = info.groupby('City')['Marks'].mean()

print(avg_marks)
````


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
import pandas as pd

info = pd.DataFrame({
    'City':['Bhopal','Bhopal','Delhi','Bhopal','Delhi'],
    'Marks':[80,90,70,85,70],
    'Age':[21,22,20,23,24]
})
filtered = info[(info['Marks'] > 70) & (info['Age'] > 20)]
print(filtered)
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
info = pd.DataFrame({
    'Employee':['Aman','Komal','Deepak','Abhishek'],
    'Departement': ['HR','IT','Manager','IT'],
    'Salary':[50000,60000,80000,60000]
})
total_salary = info.groupby('Departement')['Salary'].sum()
print(total_salary)
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
info = pd.DataFrame({
    'Name' : ['Deepak','Abhishek','Komal','Nancy','tanya'],
    'Age' :[21,22,23,25,18],
    'City':['Bhopal','Patna','Bhopal','Delhi','Patna'],
    'Marks':[70,79,75,88,80]

})
print(info['City'].unique())

print(info['City'].value_counts())

print(info[info['Marks'] > 80])

print(info.sort_values(by='Marks'))
````
