## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd
data = pd.DataFrame ({
    'Age':[25,18,30,22]
})
print(data.sort_values(by = 'Age'))
```

## Q2. Sort Marks in Descending Order

Input:
Marks = [70, 95, 80, 60]

Output:
95
80
70
60

```python
data = pd.DataFrame({
    'Marks':[70,95,80,60]
})
print(data.sort_values(by = 'Marks',ascending = False))
```

## Q3. Find Mean Age

Input:
Age = [20, 25, 30, 25]

25.0

```python
data = pd.DataFrame({
    'Age':[20,25,30,25]
})
data.mean(numeric_only= True)
```

## Q8. Count City Frequency

Input:
City = ['Bhopal','Indore','Bhopal','Jabalpur']

Output:
Bhopal      2
Indore      1
Jabalpur    1

```python
info = pd.DataFrame({
    'City':['Bhopal','Indore','Bhopal','Jabalpur']
})
print(info['City'].value_counts())
```

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
info = pd.DataFrame({
    'City':['A','A','B','C']
})
print(info['City'].value_counts(normalize = True)*100)
```

## Q10. Sort DataFrame by Salary

Input:
Name   Salary
Aman   30000
Raj    50000
Riya   40000

Output:
Raj
Riya
Aman

```python
info = pd.DataFrame({
    'Name':["Aman","Raj","Riya"],
    'Salary':[30000,50000,40000]
})
print(info.sort_values(by = 'Salary',ascending = False))
```

## Q11. Find Correlation

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
1.0

```python
info = pd.DataFrame({
    'Math':[10,20,30,40],
    'Science':[15,25,35,45]
})
print(info.corr(numeric_only = True))
```

## Q12. Find Covariance

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
166.67 (approx)

```python
info = pd.DataFrame({
    'Math':[10,20,30,40,],
    'Science':[15,25,35,45]
})
print(info.cov(numeric_only = True))
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
info = pd.DataFrame({
    'Name':['Deepak','Abhishek','Ankit','Pawan'],
    'Age':[21,22,16,18],
    'Score':[80,95,90,96],
    'Height':[160,170,172,166]
})
print(info.sort_values(by = 'Score',ascending = False))
print(info.mean(numeric_only = True))
print(info.median(numeric_only = True))
print(info['Score'].value_counts())
```
